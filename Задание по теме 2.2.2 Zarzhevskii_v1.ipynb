# 1) Загрузите датасет с помощью функции np.genfromtxt()

```python
import numpy as np
import pandas as pd
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
species = np.genfromtxt(url, delimiter = ',' , dtype = 'str')
species
```
